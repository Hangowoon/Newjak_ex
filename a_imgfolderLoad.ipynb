###1) train 데이터로 불러올 폴더 지정

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.utils.data as data
import torchvision.datasets as dset
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torch.autograd import Variable
```

```python
imgsize = 128
img_dir = "drive/MyDrive/Colab Notebooks/my_catdog/train/"
my_train = dset.ImageFolder(img_dir, transforms.Compose([
            transforms.Resize((imgsize, imgsize)),
            transforms.ToTensor()
            ]))

print(len(my_train))
```

###2) train 데이터 정보 출력

```python
nclass = len(my_train.classes)
print("# of classes(클래스 개수) : %d" %nclass)
print("클래스 이름 : ", my_train.classes)
print("클래스별 인덱스 : ", my_train.class_to_idx)
print("my_train 개수 : ", len(my_train))
```

###2) train 데이터 그리기

```python
for i in range(3):
  img = my_train[i][0].numpy()
  img = np.transpose(img, (1,2,0))

  plt.title("label : %d" %my_train[i][1])
  plt.imshow(img)
  plt.show()

  print(img.shape)
```

###3) 데이터 로드하기

```python
batch_size = 3
```

```python
train_loader = data.DataLoader(my_train, batch_size=batch_size,
                               shuffle=False, num_workers=2,
                               drop_last=True)
```

###4) 데이터 로더 사용하기

```python
n = 3
for i, [imgs, labels] in enumerate(train_loader):
  if  i>5:
    break 

  print("[%d]"%i)
  print("한 번에 로드되는 데이터 크기 :", len(imgs))

  x = Variable(imgs)
  y_ = Variable(labels)

  print(x.shape)
  print(y_.shape)

  for j in range(n):
    img = imgs[j].numpy()
    img = np.transpose(img, (1,2,0))

    plt.subplot(1, n, j+1)
    plt.imshow(img)
    plt.title("label : %d" %labels[j])
  plt.show()
```
